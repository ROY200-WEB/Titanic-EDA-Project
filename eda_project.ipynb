```python
sns.countplot(x='Survived', data=df)

plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)

plt.title("Male vs Female Passengers")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)

plt.title("Survival Based on Gender")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Age'], bins=30, kde=True)

plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)

plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)

plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(x=df['Fare'])

plt.title("Fare Outliers")
plt.show()
```

```python
sns.boxplot(x=df['Age'])

plt.title("Age Outliers")
plt.show()
```

```python
df.corr(numeric_only=True)
```

# Key Observations

1. Most passengers did not survive.

2. Female passengers had a higher survival rate compared to males.

3. Third-class passengers were the majority on the ship.

4. First-class passengers had better survival chances.

5. Most passengers were between 20–40 years old.

6. Fare column contains several outliers with very high ticket prices.

7. Age column also contains some outliers.

8. Missing values were found in the Age, Cabin, and Embarked columns.

# Conclusion

This Exploratory Data Analysis (EDA) project analyzed the Titanic dataset to identify trends, distributions, and outliers.

The analysis showed that gender and passenger class strongly influenced survival rates. Female and first-class passengers had higher chances of survival.

Several visualizations were created to better understand the dataset, including count plots, histograms, and boxplots.
